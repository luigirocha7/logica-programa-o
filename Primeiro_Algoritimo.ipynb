```python
#if / else
num1 = 10
num2 = 5
if num1 > num2:
  print('o num1 é o maior numero ')
else:
  print('o num2 é o maior numero')
```

```python
# if/elif/else
num1 = 10
num2 = 10
if num1 > num2:
  print('o num1 é o maior numero')
elif num1 < num2:
  print('o num2 é o maior numero')
else:
  print('os dois são iguais')
```
